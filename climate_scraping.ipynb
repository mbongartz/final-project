```python
#won't be using the max and min temperatures for this project, but could be useful for something else!

parameters = {"temp_avg": "tavg", "temp_max": "tmax", "temp_min": "tmin", "precip": "pcp"}
```

```python
#create function to insert fips codes, parameters, time scales, and months into url

def construct_url(fips_code, parameter, time_scale, month):
    url = "http://www.ncdc.noaa.gov/cag/time-series/us/" + fips_code + "/00/" + parameter + "/" + time_scale + "/" + month + "/1895-2015.csv?base_prd=true&firstbaseyear=1901&lastbaseyear=2000"
    
    return url
```

```python
#iterate through every possible url combination and save the csvs to the parameter names
#print the names to make sure it's working

for f in fips_codes:
    for p in parameters:
        for t in time_scales:
            for m in months:
                with open( f+"_"+p+"_"+t+"_"+m+".csv","wb" ) as writefile:
                    content = requests.get( construct_url( fips_codes[f],parameters[p],time_scales[t],months[m] ) ).content
                    writefile.write( content )
                    print f,p,t,m
```
